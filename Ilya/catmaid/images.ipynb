```python
import requests
import numpy as np
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

class connection:
    def __init__(self):
        self.session = requests.Session()

    def get_image(self, z_index, x, y, zoom_level = 0, numpy=True):
        assert 0 <= x < 64 and 0 <= y < 64
        url = f"https://flyemdev.mrc-lmb.cam.ac.uk/L1-CNS-tiles/{z_index}/{zoom_level}/{y}_{x}.jpg"
        response = self.session.get(url)
        if numpy:
            img = Image.open(BytesIO(response.content))
            return np.array(img)
        return response.content

    # jpg - original format
    def save(self, path, bytes): 
        with open(path, "wb") as f:
            f.write(bytes)
```

```python
lavra_the_pirate = connection()
```

```python
image = lavra_the_pirate.get_image(1672, 21, 54)
plt.imshow(image)
plt.show()
```
