```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
q = pd.read_csv("../Datasets/Original/all-all_connectivity_matrix.csv")
all_all = q[q.columns[1:]].to_numpy()
plt.imshow(all_all > 0) # что бы увидеть все ненулевые значения
plt.show()
```

```python
totalS = 0
for h in ("da", "dd", "ad", "aa"):
    q = pd.read_csv(f"../Datasets/Original/{h}_connectivity_matrix.csv")
    g = q[q.columns[1:]].to_numpy()
    print(h, ":", sep = '')
    plt.imshow(g > 0) # g > 0 что бы увидеть все ненулевые значения
    plt.show()
    totalS += g.sum()
print("sum of ???", totalS, "is equal to sum of all-all ???", all_all.sum())
```
