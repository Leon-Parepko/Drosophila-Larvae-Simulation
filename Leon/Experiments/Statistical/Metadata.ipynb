```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter

# 1. Загружаем граф из GML файла
G = nx.read_gml("test_graph_2n.gml")

# 2. Определяем тип каждой ноды (ожидаем атрибут 'type')
node_types = nx.get_node_attributes(G, 'type')

# Задаём уникальные типы и назначаем им цвета
unique_types = list(set(node_types.values()))
color_map = {t: i for i, t in enumerate(unique_types)}  # каждому типу — индекс
colors = [color_map[node_types.get(node, "unknown")] for node in G.nodes()]

# Рисуем граф
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # фиксированный layout для повторяемости
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.Set3, node_size=500)
plt.show()

# 3. Подсчёт количества узлов каждого типа
type_counts = Counter(node_types.values())
print("Количество узлов каждого типа:")
for t, c in type_counts.items():
    print(f"  {t}: {c}")

# 4. Проверка на связность
# 4. Проверка на связность
if G.is_directed():
    if nx.is_strongly_connected(G):
        print("Граф сильно связный (1 компонента сильной связности)")
    else:
        n_strong = nx.number_strongly_connected_components(G)
        print(f"Граф НЕ сильно связный. Кол-во компонент сильной связности: {n_strong}")

    if nx.is_weakly_connected(G):
        print("Граф слабо связный (1 компонента слабой связности)")
    else:
        n_weak = nx.number_weakly_connected_components(G)
        print(f"Граф НЕ слабо связный. Кол-во компонент слабой связности: {n_weak}")
else:
    if nx.is_connected(G):
        print("Граф связный (1 компонента связности)")
    else:
        num_components = nx.number_connected_components(G)
        print(f"Граф несвязный. Кол-во компонент связности: {num_components}")
```
